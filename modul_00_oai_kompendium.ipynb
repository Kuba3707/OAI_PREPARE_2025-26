## 5. Najważniejsze metryki — ściągawka

````python
import numpy as np

# === METRYKI KLASYFIKACJI ===
from sklearn.metrics import (
 accuracy_score, balanced_accuracy_score,
 f1_score, precision_score, recall_score,
 roc_auc_score, average_precision_score,
 confusion_matrix, classification_report
)

# Przykład
y_true = [0, 1, 1, 0, 1, 0, 1, 1]
y_pred = [0, 1, 0, 0, 1, 1, 1, 1]

print("=== METRYKI KLASYFIKACJI ===")
print(f"Accuracy: {accuracy_score(y_true, y_pred):.4f}")
print(f"Balanced Accuracy: {balanced_accuracy_score(y_true, y_pred):.4f}")
print(f"F1 (macro): {f1_score(y_true, y_pred, average='macro'):.4f}")
print(f"F1 (binary): {f1_score(y_true, y_pred):.4f}")
print(f"ROC AUC: {roc_auc_score(y_true, y_pred):.4f}")
print(f"\nClassification Report:\n{classification_report(y_true, y_pred)}")
````

````python
# === METRYKI REGRESJI / OBRAZÓW ===

def psnr(original, reconstructed, max_val=1.0):
 """Peak Signal-to-Noise Ratio — używane w odszumianiu / inpainting."""
 mse = np.mean((original - reconstructed) ** 2)
 if mse == 0:
 return float('inf')
 return 20 * np.log10(max_val / np.sqrt(mse))

def ssim_simple(img1, img2):
 """Uproszczone SSIM (Structural Similarity Index)."""
 C1 = 0.01 ** 2
 C2 = 0.03 ** 2
 mu1, mu2 = np.mean(img1), np.mean(img2)
 sigma1_sq = np.var(img1)
 sigma2_sq = np.var(img2)
 sigma12 = np.mean((img1 - mu1) * (img2 - mu2))
 return ((2*mu1*mu2 + C1) * (2*sigma12 + C2)) / ((mu1**2 + mu2**2 + C1) * (sigma1_sq + sigma2_sq + C2))

# Przykład
original = np.random.rand(32, 32)
noisy = original + 0.1 * np.random.randn(32, 32)
print(f"=== METRYKI OBRAZÓW ===")
print(f"PSNR: {psnr(original, noisy):.2f} dB")
print(f"SSIM: {ssim_simple(original, noisy):.4f}")
print(f"MSE: {np.mean((original - noisy)**2):.6f}")
````

````python
# === METRYKI NLP ===

def mean_reciprocal_rank(true_answers, predicted_rankings, K=20):
 """MRR@K — używane w zadaniu Zagadki (I OAI)."""
 rankings = []
 for true, preds in zip(true_answers, predicted_rankings):
 for rank, pred in enumerate(preds[:K], 1):
 if pred == true:
 rankings.append(1.0 / rank)
 break
 else:
 rankings.append(0.0)
 return np.mean(rankings)

# BLEU — używane w tłumaczeniu maszynowym
# from nltk.translate.bleu_score import sentence_bleu, corpus_bleu

# IoU / mIoU — używane w segmentacji (III OAI)
def iou_score(pred_mask, true_mask, num_classes=4):
 """Intersection over Union per class."""
 ious = []
 for cls in range(num_classes):
 intersection = np.sum((pred_mask == cls) & (true_mask == cls))
 union = np.sum((pred_mask == cls) | (true_mask == cls))
 if union > 0:
 ious.append(intersection / union)
 return np.mean(ious)

print("Metryki gotowe do użycia!")
````
