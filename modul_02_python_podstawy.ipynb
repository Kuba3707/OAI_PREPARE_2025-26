## 1. Typy danych i zmienne

W Pythonie **wszystko jest obiektem**. Każdy obiekt ma swój typ. Najważniejsze typy podstawowe:

| Typ | Opis | Przykłady |
|-----|------|-----------|
| `int` | Liczba całkowita | `42`, `-7`, `0` |
| `float` | Liczba zmiennoprzecinkowa | `3.14`, `-0.001` |
| `str` | Łańcuch znaków (tekst) | `'hello'`, `"world"` |
| `bool` | Wartość logiczna | `True`, `False` |
| `NoneType` | Brak wartości | `None` |

**Zmienna** to po prostu nazwa przypisana do obiektu. Python jest **dynamicznie typowany** — nie trzeba deklarować typu zmiennej.

````python
# Przykłady typów danych
wiek = 17 # int
srednia = 4.75 # float
imie = "Anna" # str
czy_uczy_ai = True # bool
wynik = None # NoneType

# Sprawdzanie typu
print(type(wiek)) # <class 'int'>
print(type(srednia)) # <class 'float'>
print(type(imie)) # <class 'str'>
print(type(czy_uczy_ai)) # <class 'bool'>
````

### Konwersja typów (rzutowanie)

Możemy zamieniać jeden typ na inny:

````python
# Konwersje typów
x = "42"
print(type(x)) # str

y = int(x) # str -> int
print(y + 8) # 50

z = float("3.14") # str -> float
print(z) # 3.14

print(str(100)) # int -> str: '100'
print(bool(0)) # int -> bool: False (0 = False, reszta = True)
print(bool(1)) # True
print(bool("")) # pusty string = False
print(bool("coś")) # niepusty string = True
````

### f-stringi — formatowanie tekstu

Sposób na wygodne wstawianie zmiennych do tekstu (od Pythona 3.6):

````python
imie = "Kuba"
wiek = 17
srednia = 4.857142857

# f-string — wstawiamy zmienne w {}
print(f"Cześć, nazywam się {imie} i mam {wiek} lat.")

# Formatowanie liczb
print(f"Moja średnia to {srednia:.2f}") # 2 miejsca po przecinku

# Wyrażenia w f-stringach
print(f"Za 5 lat będę mieć {wiek + 5} lat.")
print(f"Imię wielkimi literami: {imie.upper()}")
````

## 2. Operatory

### Operatory arytmetyczne
| Operator | Opis | Przykład |
|----------|------|----------|
| `+` | dodawanie | `5 + 3 = 8` |
| `-` | odejmowanie | `5 - 3 = 2` |
| `*` | mnożenie | `5 * 3 = 15` |
| `/` | dzielenie (float) | `7 / 2 = 3.5` |
| `//` | dzielenie całkowite | `7 // 2 = 3` |
| `%` | reszta z dzielenia (modulo) | `7 % 2 = 1` |
| `**` | potęgowanie | `2 ** 3 = 8` |

### Operatory porównania
| Operator | Opis |
|----------|------|
| `==` | równe |
| `!=` | różne |
| `<`, `>` | mniejsze, większe |
| `<=`, `>=` | mniejsze/większe lub równe |

### Operatory logiczne
| Operator | Opis |
|----------|------|
| `and` | koniunkcja (oba True) |
| `or` | alternatywa (przynajmniej jeden True) |
| `not` | negacja |

````python
# Operatory arytmetyczne
print(10 / 3) # 3.3333...
print(10 // 3) # 3 (dzielenie całkowite)
print(10 % 3) # 1 (reszta z dzielenia)
print(2 ** 10) # 1024

# Operatory porównania — zwracają bool
print(5 > 3) # True
print(5 == 5.0) # True (porównanie wartości, nie typu!)

# Operatory logiczne
print(True and False) # False
print(True or False) # True
print(not True) # False

# Praktyczne użycie
wiek = 17
print(wiek >= 16 and wiek <= 19) # True — czy jest nastolatkiem?
````

## 3. Struktury danych

### Lista (`list`) — uporządkowana, zmienna kolekcja

Lista to **najczęściej używana** struktura danych w Pythonie. Przechowuje elementy w kolejności i pozwala je modyfikować.

````python
# Tworzenie listy
oceny = [5, 4, 3, 5, 4, 5]
mieszana = [1, "dwa", 3.0, True] # może zawierać różne typy

# Indeksowanie (od 0!)
print(oceny[0]) # 5 (pierwszy element)
print(oceny[-1]) # 5 (ostatni element)
print(oceny[1:4]) # [4, 3, 5] — wycinanie (slice): od indeksu 1 do 3

# Modyfikacja
oceny.append(6) # dodaj na koniec
oceny.insert(0, 2) # wstaw 2 na pozycję 0
oceny.remove(3) # usuń pierwszą wartość 3
ostatni = oceny.pop() # usuń i zwróć ostatni element

# Przydatne operacje
print(len(oceny)) # długość
print(sum(oceny)) # suma
print(sorted(oceny)) # posortowana kopia
print(3 in oceny) # czy 3 jest w liście?
print(min(oceny), max(oceny)) # minimum i maximum
````

### Słownik (`dict`) — pary klucz: wartość

Słowniki to **mapowania** — szybki dostęp do wartości po kluczu (nie po indeksie).

````python
# Tworzenie słownika
uczen = {
 "imie": "Anna",
 "wiek": 17,
 "oceny": [5, 4, 5, 3],
 "olimpiada": True
}

# Dostęp do wartości
print(uczen["imie"]) # Anna
print(uczen.get("klasa", "brak")) # "brak" — bezpieczne pobieranie

# Modyfikacja
uczen["klasa"] = "2c" # dodanie nowego klucza
uczen["wiek"] = 18 # zmiana istniejącej wartości

# Iteracja
for klucz, wartosc in uczen.items():
 print(f"{klucz}: {wartosc}")

# Przydatne metody
print(uczen.keys()) # wszystkie klucze
print(uczen.values()) # wszystkie wartości
print("imie" in uczen) # True — czy klucz istnieje?
````

### Zbiór (`set`) — nieuporządkowana kolekcja unikalnych elementów

Zbiory **automatycznie usuwają duplikaty** i umożliwiają szybkie operacje matematyczne na zbiorach.

````python
# Tworzenie zbioru
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# Duplikaty są automatycznie usuwane
C = {1, 1, 2, 2, 3}
print(C) # {1, 2, 3}

# Operacje na zbiorach
print(A | B) # suma (union): {1, 2, 3, 4, 5, 6, 7, 8}
print(A & B) # przecięcie (intersection): {4, 5}
print(A - B) # różnica: {1, 2, 3}

# Tworzenie zbioru z listy — usuwanie duplikatów
lista = [1, 2, 2, 3, 3, 3]
unikalne = list(set(lista)) # [1, 2, 3]
print(unikalne)
````

### Krotka (`tuple`) — jak lista, ale niezmienna

Krotek nie można modyfikować po utworzeniu. Używa się ich gdy chcemy zapewnić, że dane się nie zmienią.

````python
# Krotki
punkt = (3, 7)
rgb = (255, 128, 0)

# Dostęp jak do listy
print(punkt[0]) # 3

# Rozpakowywanie
x, y = punkt
print(f"x={x}, y={y}")

r, g, b = rgb
print(f"R={r}, G={g}, B={b}")

# UWAGA: nie można modyfikować!
# punkt[0] = 5 # TypeError!
````

## 4. Instrukcje warunkowe i pętle

### `if / elif / else`

````python
wynik = 75

if wynik >= 90:
 ocena = "celujący"
elif wynik >= 75:
 ocena = "bardzo dobry"
elif wynik >= 50:
 ocena = "dobry"
elif wynik >= 30:
 ocena = "dostateczny"
else:
 ocena = "niedostateczny"

print(f"Wynik {wynik}% → ocena: {ocena}")
````

### Pętla `for` — iteracja po elementach

````python
# Iteracja po liście
owoce = ["jabłko", "banan", "wiśnia"]
for owoc in owoce:
 print(owoc)

# range() — generuje ciąg liczb
for i in range(5): # 0, 1, 2, 3, 4
 print(i, end=" ")
print()

for i in range(2, 10, 3): # start=2, stop=10, krok=3 → 2, 5, 8
 print(i, end=" ")
print()

# enumerate() — indeks + element
for i, owoc in enumerate(owoce):
 print(f"{i}: {owoc}")

# zip() — łączenie dwóch list
imiona = ["Anna", "Jan", "Maria"]
wyniki = [95, 78, 88]
for imie, wynik in zip(imiona, wyniki):
 print(f"{imie}: {wynik}%")
````

### Pętla `while` — powtarzaj dopóki warunek jest prawdziwy

````python
# Przykład: szukamy pierwszej potęgi 2 większej od 1000
n = 1
while n <= 1000:
 n *= 2
print(f"Pierwsza potęga 2 > 1000: {n}") # 1024

# break i continue
for i in range(10):
 if i == 3:
 continue # pomiń 3
 if i == 7:
 break # przerwij pętlę
 print(i, end=" ") # 0 1 2 4 5 6
````

## 5. Funkcje

Funkcje pozwalają opakowywać logikę do wielokrotnego użytku.

````python
# Definiowanie funkcji
def srednia(lista):
 """Oblicza średnią arytmetyczną listy liczb."""
 return sum(lista) / len(lista)

oceny = [5, 4, 3, 5, 4]
print(f"Średnia: {srednia(oceny)}") # 4.2

# Argumenty domyślne
def potega(baza, wykladnik=2):
 return baza ** wykladnik

print(potega(3)) # 9 (domyślnie kwadrat)
print(potega(2, 10)) # 1024

# Zwracanie wielu wartości (jako krotka)
def statystyki(lista):
 return min(lista), max(lista), srednia(lista)

mi, ma, sr = statystyki(oceny)
print(f"Min: {mi}, Max: {ma}, Średnia: {sr}")
````

### Funkcje lambda — jednolinijkowe funkcje anonimowe

````python
# Lambda — krótka funkcja bez nazwy
kwadrat = lambda x: x ** 2
print(kwadrat(5)) # 25

# Przydatne z sorted(), map(), filter()
uczniowie = [("Anna", 95), ("Jan", 78), ("Maria", 88)]

# Sortowanie po wyniku (drugi element krotki)
posortowani = sorted(uczniowie, key=lambda u: u[1], reverse=True)
print(posortowani) # [(Anna, 95), (Maria, 88), (Jan, 78)]

# map() — zastosuj funkcję do każdego elementu
liczby = [1, 2, 3, 4, 5]
kwadraty = list(map(lambda x: x**2, liczby))
print(kwadraty) # [1, 4, 9, 16, 25]

# filter() — filtruj elementy
parzyste = list(filter(lambda x: x % 2 == 0, liczby))
print(parzyste) # [2, 4]
````

## 6. Klasy i obiekty

Klasy pozwalają tworzyć własne typy danych z atrybutami (danymi) i metodami (funkcjami). To jest **kluczowe** w PyTorch — modele sieci neuronowych to klasy!

````python
class Uczen:
 def __init__(self, imie, klasa, oceny=None):
 """Konstruktor — wywoływany przy tworzeniu obiektu."""
 self.imie = imie
 self.klasa = klasa
 self.oceny = oceny if oceny is not None else []

 def dodaj_ocene(self, ocena):
 self.oceny.append(ocena)

 def srednia(self):
 if not self.oceny:
 return 0.0
 return sum(self.oceny) / len(self.oceny)

 def __str__(self):
 """Reprezentacja tekstowa (wywoływana przez print())."""
 return f"{self.imie} ({self.klasa}) — średnia: {self.srednia():.2f}"


# Tworzenie obiektów
anna = Uczen("Anna", "2c", [5, 4, 5])
jan = Uczen("Jan", "2c")
jan.dodaj_ocene(4)
jan.dodaj_ocene(3)

print(anna) # Anna (2c) — średnia: 4.67
print(jan) # Jan (2c) — średnia: 3.50
print(anna.oceny) # [5, 4, 5]
````

### Dziedziczenie

Klasy mogą dziedziczyć po innych klasach. W PyTorch **każdy model dziedziczy po `nn.Module`**.

````python
class OlimpijczykAI(Uczen):
 """Klasa dziedzicząca po Uczen."""
 def __init__(self, imie, klasa, oceny=None, etap_olimpiady=1):
 super().__init__(imie, klasa, oceny) # wywołanie konstruktora rodzica
 self.etap_olimpiady = etap_olimpiady

 def __str__(self):
 bazowy = super().__str__()
 return f"{bazowy} | OAI etap {self.etap_olimpiady}"


olimpijczyk = OlimpijczykAI("Maria", "3a", [5, 5, 5], etap_olimpiady=2)
print(olimpijczyk) # Maria (3a) — średnia: 5.00 | OAI etap 2
````

## 7. List/Dict Comprehension

Skrócony sposób tworzenia list i słowników — bardzo popularny w Pythonie.

````python
# List comprehension
kwadraty = [x**2 for x in range(10)]
print(kwadraty) # [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

# Z warunkiem (filtracja)
parzyste_kwadraty = [x**2 for x in range(10) if x % 2 == 0]
print(parzyste_kwadraty) # [0, 4, 16, 36, 64]

# Dict comprehension
kwadraty_dict = {x: x**2 for x in range(6)}
print(kwadraty_dict) # {0: 0, 1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# Praktyczny przykład: zamiana listy krotek na słownik
dane = [("Anna", 95), ("Jan", 78), ("Maria", 88)]
wyniki_dict = {imie: wynik for imie, wynik in dane}
print(wyniki_dict) # {'Anna': 95, 'Jan': 78, 'Maria': 88}
````

## 8. Złożoność obliczeniowa (Big-O)

Na zawodach ważne jest, aby rozwiązanie **zmieściło się w limicie czasu**. Złożoność obliczeniowa opisuje, jak czas (lub pamięć) rośnie wraz z rozmiarem danych $n$.

### Notacja Big-O

$O(f(n))$ — **asymptotyczne górne ograniczenie** czasu działania:

$$T(n) = O(f(n)) \iff \exists\, c > 0,\, n_0 \in \mathbb{N}: \forall\, n \geq n_0,\; T(n) \leq c \cdot f(n)$$

### Najważniejsze klasy złożoności

| Złożoność | Nazwa | Przykład | $n = 10^6$ |
|-----------|-------|----------|------------|
| $O(1)$ | stała | dostęp do elementu tablicy | natychmiast |
| $O(\log n)$ | logarytmiczna | binary search | ~20 operacji |
| $O(n)$ | liniowa | przejście tablicy | $10^6$ |
| $O(n \log n)$ | liniowo-logarytmiczna | sortowanie (merge sort) | $\sim 2 \cdot 10^7$ |
| $O(n^2)$ | kwadratowa | dwie zagnieżdżone pętle | $10^{12}$ — za wolno! |
| $O(2^n)$ | wykładnicza | brute-force podzbiorów | nierealne |

### Reguła kciuka (zawody)

Typowy komputer wykonuje $\sim 10^8$ operacji/s. Jeśli $n \leq 10^6$, algorytm $O(n \log n)$ zmieści się w 1s, ale $O(n^2)$ nie.

### Złożoność operacji w Pythonie

| Struktura | Operacja | Złożoność |
|-----------|----------|-----------|
| `list` | `append()` | $O(1)$ amortyzowane |
| `list` | `x in list` | $O(n)$ |
| `list` | `sort()` | $O(n \log n)$ — Timsort |
| `dict` | `d[key]`, `key in d` | $O(1)$ średnio |
| `set` | `x in set`, `add()` | $O(1)$ średnio |

> **Tip na zawody**: Jeśli potrzebujesz szybkiego wyszukiwania ($O(1)$), użyj `set` lub `dict` zamiast `list`!
